```python

# Create a DataFrame with some missing values
data = {
    'age': [25, 30, np.nan, 40, 35, np.nan, 28],
    'income': [50000, 60000, 70000, np.nan, 85000, 65000, 72000],
    'education': ['Bachelor', 'Master', 'PhD', 'Bachelor', np.nan, 'Master', 'PhD'],
    'city': ['New York', 'London', 'Tokyo', 'Paris', 'New York', np.nan, 'London']
}

dummy_df = pd.DataFrame(data)

# Display the DataFrame with missing values
print("DataFrame with Missing Values:")
print(dummy_df)
```

```python
dummy_df['age'].fillna(dummy_df['age'].mean(),inplace=True)
dummy_df
```

```python
income_modif=['income']
dummy_df['income'].fillna(dummy_df['income'].mean(),inplace=True)
dummy_df
```

```python
category_col=['education','city']
for col in category_col:
  dummy_df[col]=dummy_df[col].fillna(dummy_df[col].mode()[0])
```

```python
dummy_df
```

```python
Q1=dummy_df['age'].quantile(0.25)
Q3=dummy_df['age'].quantile(0.75)
IQR=Q3-Q1

max=Q3 + (1.5*IQR)
min=Q1 - (1.5*IQR)

kondisi_lower_than = dummy_df['age'] < min
kondisi_more_than = dummy_df['age'] > max

dummy_df.mask(cond=kondisi_lower_than, other=max, axis=0, inplace=True)
dummy_df.mask(cond=kondisi_lower_than, other=min, axis=0, inplace=True)
```

```python
import numpy as np
import pandas as pd

# Create a DataFrame with outliers
data = {
    'value': [10, 12, 15, 11, 13, 14, 1000],  # 1000 is an outlier
    'value2': [10, 12, 65, 61, 18, 12, 24]  # 1000 is an outlier
}
df = pd.DataFrame(data)  # Convert the dictionary to a DataFrame

# Now you can use quantile on the DataFrame columns
Q1 = df['value'].quantile(0.25)
Q3 = df['value'].quantile(0.75)
IQR = Q3 - Q1

max = Q3 + (1.5 * IQR)
min = Q1 - (1.5 * IQR)

kondisi_lower_than = df['value'] < min  # Use df['value']
kondisi_more_than = df['value'] > max  # Use df['value']

# ini mempengaruhi semua kolom values
df.mask(cond=kondisi_lower_than, other=max, axis=0, inplace=True)  # Apply mask to df
df.mask(cond=kondisi_lower_than, other=min, axis=0, inplace=True)  # Apply mask to df
df
```

```python
df['value']= df['value'].mask(kondisi_lower_than,min)
df['value']= df['value'].mask(kondisi_more_than,max)
df

# ini mempengaruhi kolom value doang
```
